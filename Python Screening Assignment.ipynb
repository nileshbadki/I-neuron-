Suppose that we need to develop a payroll program for a college Teachers. The college Teachers has two groups: full-time Teachers and hourly base Teachers. The full-time Teachers get a fixed salary while the hourly base employees get paid by hourly wages for their services. To model the payroll program in an object-oriented way, you may come up with the following classes: Teachers, FulltimeTeachers, HourlybaseTeachers, and Payroll

#### The Teachers class

The Teachers class represents an teachers, either full-time or hourly. The Teachers class should be an abstract class because thereâ€™re only full-time employees and hourly employees, no general employees exist.

```python
from abc import ABC, abstractmethod


class Teachers(ABC):
    def __init__(self, first_name, last_name):
        self.first_name = first_name
        self.last_name = last_name

    @property
    
    def full_name(self):
        return f"{self.first_name} {self.last_name}"

    @abstractmethod
    
    def get_salary(self):
        pass
```

```python
# The FulltimeTeachers class inherits from the Teachers class.

class FulltimeTeachers(Teachers):
    def __init__(self, first_name, last_name, salary):
        super().__init__(first_name, last_name)
        self.salary = salary

    def get_salary(self):
        return self.salary
```

```python
# The HourlybaseTeachers also inherits from the Teachers class.

class HourlybaseTeachers(Teachers):
    def __init__(self, first_name, last_name, worked_hours, rate):
        super().__init__(first_name, last_name)
        self.worked_hours = worked_hours
        self.rate = rate

    def get_salary(self):
        return self.worked_hours * self.rate
```

```python
class Payroll:
    def __init__(self):
        self.teachers_list = []

    def add(self, teachers):
        self.teachers_list.append(teachers)

    def print(self):
        for i in self.teachers_list:
            print(f"{i.full_name} \t ${i.get_salary()}")
```
